# K-means demo
## Build your model

**Step1**: Load in the dataset

```python
import pandas as pd

df = pd.read_csv('data/k-means-example.csv')
points = df.values

points[0:3]
```

**Step2 **: Generate model and predict labels
- Use the `KMeans` method from `sklearn` to generate your k-means model. 
- Use the `.fit()` method to fit the model to your data. 
- Finally, use the `.predict()` method of `model` to predict the cluster labels for your data.

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(points)
labels = model.predict(points)
```

**Step 3**: Visualize your clusters

```python
import matplotlib.pyplot as plt

xs = points[:,0]
ys = points[:,1]

plt.scatter(xs, ys, c=labels)
plt.show()
```

Great! You have used k-means to cluster your data! One step futher: plot the centroids on this scatter plot.

**Step 4**: Get the centroid coordinates using the `.cluster_centers_` attribude of `model`. Assign these coordinates to `centroids`.

```python
centroids = model.cluster_centers_
centroids[0:3]
```

Looks like we have x,y points again.   
**Step 5**: Assign column 0 of `centroids` to `centroids_x` and column 1 to `centroids_y`.

```python
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
```

**Step 6**: Overlay your scatter plot from the original labeled data with a plot of the centroid positions.

```python
plt.scatter(xs, ys, c=labels)
plt.scatter(centroids_x, centroids_y, marker='X', s=200)
plt.show()
```

**Complete!** Centroid are important because they enable k-means to assign new, previously unseen points to existing clusters.

## Adding new points to your model

```python
df = pd.read_csv('data/k-means-new-points.csv')
new_points = df.values

new_points[0:3]
```

```python
new_labels = model.predict(new_points)
new_labels[0:3]
```

```python
new_xs = new_points[:,0]
new_ys = new_points[:,1]

#plt.scatter(xs, ys, c=labels)
plt.scatter(new_xs, new_ys, c=new_labels)
plt.scatter(centroids_x, centroids_y, marker='X', s=200)

plt.show()
```
